```python
aggdict = {
    'Open':'first',
    'High':'max',
    'Low':'min',
    'Close':'last',
    'volume':'sum'
}
```

```python
 def HigherTimeframeMAs(source , newFreq ='5min'):
        df = source.groupby(pd.Grouper(level=0, freq= newFreq, label='left')).agg(aggdict)
```
