```python
import pandas as pd
import datetime as dt
```

### Read in data
Data source: Tradestation export

```python
#Read from .txt in Tradestation format
df = pd.read_csv("SPY_Daily.txt")

#Rename columns
df = df.rename(columns={
    'Open': 'open',
    'High': 'high',
    'Low': 'low',
    'Close': 'close',
    'Vol': 'volume'})

#Convert strings of date and time to datetime format
df["datetime"] = df["Date"] + " " + df["Time"]
df["datetime"] =pd.to_datetime(df["datetime"])

#Shift 1min to the left for proper higher timeframe generation
df["datetime"] = df["datetime"] - dt.timedelta(minutes=1)
df= df.set_index('datetime')

#drop unneccessary columns
df = df.drop(columns=['OI'])
df
```

### Add column for weekday

```python
#Add weekday
df = df.reset_index()
df['weekdayInt'] = df['datetime'].dt.weekday

#If you prefer to have the days of the week intstead of int representation
df.loc[df['weekdayInt'] == 0, 'weekday'] = "Monday"
df.loc[df['weekdayInt'] == 1, 'weekday'] = "Tuesday"
df.loc[df['weekdayInt'] == 2, 'weekday'] = "Wednesday"
df.loc[df['weekdayInt'] == 3, 'weekday'] = "Thursday"
df.loc[df['weekdayInt'] == 4, 'weekday'] = "Friday"
df.loc[df['weekdayInt'] == 5, 'weekday'] = "Saturday"
df.loc[df['weekdayInt'] == 6, 'weekday'] = "Sunday"

df = df.set_index('datetime')
df
```
